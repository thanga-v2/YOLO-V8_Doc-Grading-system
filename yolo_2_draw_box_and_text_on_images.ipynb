```python
import cv2

img = cv2.imread('cars.jpg')

# Simulate what YOLO would return
detections = [
    {'label': 'person',    'conf': 0.95, 'box': (50, 50, 250, 400)},
    {'label': 'signature', 'conf': 0.88, 'box': (300, 200, 500, 280)},
    {'label': 'stamp',     'conf': 0.76, 'box': (400, 350, 550, 450)},
]

for det in detections:
    x1, y1, x2, y2 = det['box']
    label = det['label']
    conf = det['conf']

    # Draw box
    cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)

    # Draw label background (so text is readable)
    label_text = f"{label} {conf:.2f}"
    (text_w, text_h), _ = cv2.getTextSize(
        label_text, cv2.FONT_HERSHEY_SIMPLEX, 0.6, 2
    )
    cv2.rectangle(img,
        (x1, y1 - text_h - 8),
        (x1 + text_w, y1),
        (0, 255, 0), -1)   # filled background

    # Draw label text (black on green background)
    cv2.putText(img, label_text,
        (x1, y1 - 5),
        cv2.FONT_HERSHEY_SIMPLEX,
        0.6, (0, 0, 0), 2)

cv2_imshow( img)
```

```python


# Load image or create blank
img = cv2.imread('person.jpg')
if img is None:
    img = np.ones((600, 800, 3), dtype=np.uint8) * 200

# Draw 3 detection boxes like YOLO would
boxes = [
    ((30, 30, 250, 350),   'document',  0.97, (0, 255, 0)),
    ((270, 50, 450, 200),  'signature', 0.89, (0, 165, 255)),
    ((500, 300, 700, 420), 'stamp',     0.73, (0, 0, 255)),
]

for (x1,y1,x2,y2), label, conf, color in boxes:
    # Bounding box
    cv2.rectangle(img, (x1,y1), (x2,y2), color, 2)

    # Label with background
    text = f'{label} {conf:.0%}'
    (tw, th), _ = cv2.getTextSize(text, cv2.FONT_HERSHEY_SIMPLEX, 0.65, 2)
    cv2.rectangle(img, (x1, y1-th-10), (x1+tw+4, y1), color, -1)
    cv2.putText(img, text, (x1+2, y1-5),
        cv2.FONT_HERSHEY_SIMPLEX, 0.65, (255,255,255), 2)

cv2_imshow( img)
cv2.imwrite('day2_output.jpg', img)
print("Saved day2_output.jpg")
```
