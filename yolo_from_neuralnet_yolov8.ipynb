```python
from ultralytics import YOLO

# Load pretrained model (downloads automatically first time ~6MB)
model = YOLO('yolov8n.pt')   # n = nano (smallest, fastest)

# Run on image
results = model('person.jpg')

# That's it! Results contains everything.
print(type(results))         # list
print(type(results[0]))      # ultralytics Results object
```

```python
model = YOLO('yolov8n.pt')
results = model('person.jpg')

result = results[0]   # first (and only) image result

# How many objects detected?
print(f"Objects detected: {len(result.boxes)}")

for box in result.boxes:
    # Class ID (number)
    class_id = int(box.cls[0])

    # Class name (string)
    label = result.names[class_id]

    # Confidence score (0.0 to 1.0)
    confidence = float(box.conf[0])

    # Bounding box coordinates
    x1, y1, x2, y2 = box.xyxy[0].tolist()

    print(f"{label}: {confidence:.2f} at [{x1:.0f},{y1:.0f},{x2:.0f},{y2:.0f}]")
```

```python
from ultralytics import YOLO

model = YOLO('yolov8n.pt')

# predict() saves result image automatically
results = model.predict(
    source='person.jpg',
    save=True,           # save annotated image
    conf=0.5,            # minimum confidence threshold
    project='runs',      # save folder
    name='day5'          # subfolder name
)

# Image saved to: runs/day5/photo.jpg
print("Saved to runs/day5/")
```
